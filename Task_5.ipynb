```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load the dataset
df = pd.read_csv('train.csv')

# Display first few rows
df.head()
```

Basic Info & Descriptive Stats

```python
# DataFrame structure
df.info()

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()

# Count unique values per column
df.nunique()
```

Univariate Analysis

```python
#Target Variable: Survived
```

```python
# Survival distribution
df['Survived'].value_counts().plot(kind='bar', color=['salmon', 'skyblue'])
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
#ategorical Variables
```

```python
# Pclass
sns.countplot(x='Pclass', data=df, palette='pastel')
plt.title('Passenger Class Distribution')
plt.show()

# Sex
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title('Gender Distribution')
plt.show()

# Embarked
sns.countplot(x='Embarked', data=df, palette='coolwarm')
plt.title('Port of Embarkation')
plt.show()
```

```python
#Numerical Variables
```

```python
# Age distribution
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.show()

# Fare distribution
sns.histplot(df['Fare'], bins=30, kde=True, color='green')
plt.title('Fare Distribution')
plt.show()
```

Bivariate Analysis

```python
#Survival by Gender
```

```python
sns.barplot(x='Sex', y='Survived', data=df, palette='Set2')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
#Survival by Class
```

```python
sns.barplot(x='Pclass', y='Survived', data=df, palette='pastel')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
#Survival by Embarkation Port
```

```python
sns.barplot(x='Embarked', y='Survived', data=df, palette='coolwarm')
plt.title('Survival Rate by Embarked Location')
plt.show()
```

```python
#Boxplots for Numerical Features
```

```python
# Age vs Survived
sns.boxplot(x='Survived', y='Age', data=df, palette='Set3')
plt.title('Age Distribution by Survival')
plt.show()

# Fare vs Survived
sns.boxplot(x='Survived', y='Fare', data=df, palette='Set1')
plt.title('Fare Distribution by Survival')
plt.show()
```

Relationships Between Features

```python
#Correlation Heatmap
```

```python
# Compute correlations
corr_matrix = df.corr(numeric_only=True)

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Pairplot
```

```python
# Select relevant columns
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='husl')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```
